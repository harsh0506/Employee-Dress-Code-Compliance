```python
import os
from collections import Counter

# Set the path to the labels folder
labels_folder = '/content/train/labels'

# Initialize a counter to keep track of the class counts
class_counter = Counter()

# Loop through each label file in the folder
for label_file in os.listdir(labels_folder):
    if label_file.endswith('.txt'):  # Check if the file is a text file (label file)
        file_path = os.path.join(labels_folder, label_file)

        # Open and read the label file
        with open(file_path, 'r') as file:
            lines = file.readlines()

            # Extract the class ID from each line (assuming the first value is the class ID)
            for line in lines:
                class_id = line.split()[0]  # Get the class ID (first value in the line)
                class_counter[class_id] += 1

# Display the number of classes and their respective counts
num_classes = len(class_counter)
print(f'Number of classes: {num_classes}')
print('Class counts:')
for class_id, count in class_counter.items():
    print(f'Class {class_id}: {count} instances')
```
