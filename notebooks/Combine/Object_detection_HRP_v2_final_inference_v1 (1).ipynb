```python
import torch
import torchvision.transforms as transforms
from PIL import Image
import requests
from io import BytesIO
import timm
import torch.nn as nn
import numpy as np

class EfficientNetV2B0(nn.Module):
    def __init__(self, num_classes):
        super(EfficientNetV2B0, self).__init__()
        self.model = timm.create_model('tf_efficientnetv2_b2.in1k', pretrained=True)
        num_ftrs = self.model.get_classifier().in_features
        self.model.classifier = nn.Sequential(
            nn.Linear(num_ftrs, num_classes),
            nn.Dropout(0.5)
        )

    def forward(self, x):
        return self.model(x)

def load_model(model_path, num_classes, device='cpu'):
    """
    Load the trained model
    """
    model = EfficientNetV2B0(num_classes=8)
    model.load_state_dict(torch.load(model_path, map_location=device))
    model = model.to(device)
    model.eval()
    return model

def process_image(image, img_size=(224, 224)):
    """
    Process image for model input
    """
    transform = transforms.Compose([
        transforms.Resize(img_size),
        transforms.CenterCrop(img_size),
        transforms.ToTensor(),
    ])
    return transform(image).unsqueeze(0)

def predict_image(model, image_input, class_names, device='cpu'):
    """
    Make prediction for a given image input

    Parameters:
    -----------
    model : torch.nn.Module
        Trained PyTorch model
    image_input : str or PIL.Image or numpy.ndarray
        Can be:
        - Path to image file
        - URL to image
        - PIL Image object
        - Numpy array
    class_names : list
        List of class names
    device : str
        'cpu' or 'cuda' (default: 'cpu')

    Returns:
    --------
    dict
        Dictionary containing predicted class and probabilities
    """
    try:
        # Handle different input types
        if isinstance(image_input, str):
            if image_input.startswith(('http://', 'https://')):
                # Handle URL
                response = requests.get(image_input)
                image = Image.open(BytesIO(response.content)).convert('RGB')
            else:
                # Handle local file path
                image = Image.open(image_input).convert('RGB')
        elif isinstance(image_input, np.ndarray):
            # Handle numpy array
            image = Image.fromarray(np.uint8(image_input)).convert('RGB')
        elif isinstance(image_input, Image.Image):
            # Handle PIL Image
            image = image_input.convert('RGB')
        else:
            raise ValueError("Unsupported input type. Please provide either an image path, URL, PIL Image, or numpy array")

        # Process image
        img_tensor = process_image(image).to(device)

        # Get prediction
        with torch.no_grad():
            outputs = model(img_tensor)
            probabilities = torch.nn.functional.softmax(outputs, dim=1)

            # Get top prediction
            _, predicted = torch.max(outputs.data, 1)

            # Get all probabilities
            probs = probabilities.cpu().numpy()[0]

            # Create results dictionary
            results = {
                'predicted_class': class_names[predicted.item()],
                'confidence': float(probs[predicted.item()]),
                'probabilities': {class_names[i]: float(prob)
                                for i, prob in enumerate(probs)}
            }

            return results

    except Exception as e:
        return {'error': str(e)}

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, random_split
from torchvision import datasets, transforms
import timm

# Parameters
img_size = (224, 224)
data_dir = '/content/drive/MyDrive/Clothing_Data_HRP/classification data'

# Data Augmentation
transform = transforms.Compose([
    transforms.Resize(img_size),
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(15),
    transforms.ColorJitter(brightness=0.1, contrast=0.1, saturation=0.1, hue=0.05),
    transforms.RandomResizedCrop(size=img_size, scale=(0.9, 1.0)),
    transforms.ToTensor(),
])
dataset = datasets.ImageFolder(data_dir, transform=transform)

num_classes = len(dataset.classes)  # Number of classes in your dataset
class_names = dataset.classes


# Example usage function
def inference_example():
    """
    Example of how to use the inference function
    """
    # Model parameters
    model_path = '/content/drive/MyDrive/Clothing_Data_HRP/classification data/Model/final_model_b2_v1.pth'  # Path to your saved model
    data_dir = '/content/drive/MyDrive/Clothing_Data_HRP'

    # Data Augmentation
    transform = transforms.Compose([
        transforms.Resize(img_size),
        transforms.RandomHorizontalFlip(),
        transforms.RandomRotation(15),
        transforms.ColorJitter(brightness=0.1, contrast=0.1, saturation=0.1, hue=0.05),
        transforms.RandomResizedCrop(size=img_size, scale=(0.9, 1.0)),
        transforms.ToTensor(),
    ])
    dataset = datasets.ImageFolder(data_dir, transform=transform)

    num_classes = 8  # Number of classes in your dataset
    class_names = ['female_bodycon', 'female_pant_suit_top', 'female_pants_bottom', 'informal cloths men', 'informal cloths women', 'male_pants', 'shirt men', 'suits men'] # List of class names
    print(class_names)

    # Load model
    device = 'cuda' if torch.cuda.is_available() else 'cpu'
    model = load_model(model_path, 8, device)

    # Example usage with different input types

    # 1. Local image path
    result = predict_image(model, '/content/output/263155-1600x1030-professional-business-attire-women_jpg.rf.0215deaeebf1db466a9245fdd6c65065_complete_4_0.58.jpg', class_names, device)
    print("Prediction from local image:", result)

    # 2. Image URL
    '''url = "https://example.com/image.jpg"
    result = predict_image(model, url, class_names, device)
    print("Prediction from URL:", result)'''

# Function to get top N predictions
def get_top_n_predictions(model, image_path, class_names, n=3, device='cpu'):
    """
    Get top N predictions for an image
    """
    results = predict_image(model, image_path, class_names, device)
    if 'error' not in results:
        # Sort probabilities by value
        sorted_probs = sorted(results['probabilities'].items(),
                            key=lambda x: x[1],
                            reverse=True)[:n]
        return {
            'top_predictions': [
                {'class': class_name, 'probability': prob}
                for class_name, prob in sorted_probs
            ]
        }
    return results

# Example of how to use the functions
if __name__ == "__main__":
    # Example setup
    model_path = '/content/drive/MyDrive/Clothing_Data_HRP/Model/final_model_b2_v1.pth'
    num_classes = len(dataset.classes)  # Replace with your number of classes
    class_names = dataset.classes  # Replace with your class names
    print(num_classes, class_names)

    # Load model
    device = 'cuda' if torch.cuda.is_available() else 'cpu'
    print(f"Using device: {device}")

    model = load_model(model_path, num_classes, device)

    # Example image path
    image_path = '/content/output/image_1.jpg'

    # Get prediction
    result = predict_image(model, image_path, class_names, device)
    print(result)
    print("\nSingle prediction:")

    print(f"Predicted class: {result['predicted_class']}")
    print(f"Confidence: {result['confidence']:.2%}")

    # Get top 3 predictions
    top_3 = get_top_n_predictions(model, image_path, class_names, n=3, device=device)
    print("\nTop 3 predictions:")
    for pred in top_3['top_predictions']:
        print(f"{pred['class']}: {pred['probability']:.2%}")
```

```python
import json
from datetime import datetime
from pathlib import Path
import time
import os

class DetectionLogger:
    def __init__(self, log_file_path="detection_logs.json"):
        """Initialize the logger with a path for the JSON log file"""
        self.log_file_path = log_file_path
        self.current_session = {
            "session_id": datetime.now().strftime("%Y%m%d_%H%M%S"),
            "start_time": datetime.now().isoformat(),
            "operations": [],
            "summary": {}
        }

    def log_operation(self, operation_type, details):
        """Log an individual operation with timestamp"""
        operation = {
            "type": operation_type,
            "timestamp": datetime.now().isoformat(),
            "details": details
        }
        self.current_session["operations"].append(operation)

    def save_logs(self):
        """Save the current session logs to JSON file"""
        self.current_session["end_time"] = datetime.now().isoformat()

        # Load existing logs if file exists
        existing_logs = []
        if os.path.exists(self.log_file_path):
            with open(self.log_file_path, 'r') as f:
                existing_logs = json.load(f)

        # Append current session
        existing_logs.append(self.current_session)

        # Save updated logs
        with open(self.log_file_path, 'w') as f:
            json.dump(existing_logs, f, indent=2)

def load_model(model_path, logger):
    """Load the YOLO model and log the operation"""
    try:
        start_time = time.time()
        model = YOLO(model_path)
        end_time = time.time()

        logger.log_operation("model_load", {
            "model_path": str(model_path),
            "status": "success",
            "duration": end_time - start_time
        })
        return model
    except Exception as e:
        logger.log_operation("model_load", {
            "model_path": str(model_path),
            "status": "error",
            "error": str(e)
        })
        raise

def process_image(image_path, model, conf_threshold=0.25, logger=None):
    """Process a single image and log the operation"""
    try:
        start_time = time.time()
        results = model(image_path)[0]
        end_time = time.time()

        if logger:
            logger.log_operation("image_process", {
                "image_path": str(image_path),
                "conf_threshold": conf_threshold,
                "status": "success",
                "duration": end_time - start_time,
                "num_detections": len(results.boxes)
            })
        return results
    except Exception as e:
        if logger:
            logger.log_operation("image_process", {
                "image_path": str(image_path),
                "status": "error",
                "error": str(e)
            })
        raise

def crop_and_save_detections(image_path, results, output_dir, conf_threshold=0.25, logger=None):
    """Crop detected objects and save them with logging"""
    start_time = time.time()
    crops = []

    try:
        image = cv2.imread(image_path)
        if image is None:
            raise Exception(f"Could not read image: {image_path}")

        output_dir = Path(output_dir)
        output_dir.mkdir(parents=True, exist_ok=True)
        base_filename = Path(image_path).stem

        for idx, detection in enumerate(results.boxes.data):
            x1, y1, x2, y2, conf, cls = detection

            if conf < conf_threshold:
                continue

            x1, y1, x2, y2 = map(int, [x1, y1, x2, y2])
            x1, y1 = max(0, x1), max(0, y1)
            x2, y2 = min(image.shape[1], x2), min(image.shape[0], y2)

            crop = image[y1:y2, x1:x2]
            class_name = results.names[int(cls)]
            output_filename = f"{base_filename}_{class_name}_{idx}_{conf:.2f}.jpg"
            output_path = output_dir / output_filename

            cv2.imwrite(str(output_path), crop)
            crop_info = {
                'path': str(output_path),
                'confidence': float(conf),
                'class': class_name,
                'bbox': (x1, y1, x2, y2)
            }
            crops.append(crop_info)

        end_time = time.time()

        if logger:
            logger.log_operation("crop_and_save", {
                "image_path": str(image_path),
                "output_dir": str(output_dir),
                "status": "success",
                "duration": end_time - start_time,
                "num_crops": len(crops),
                "crops": crops
            })

        return crops
    except Exception as e:
        if logger:
            logger.log_operation("crop_and_save", {
                "image_path": str(image_path),
                "output_dir": str(output_dir),
                "status": "error",
                "error": str(e)
            })
        raise

def detect_and_crop_humans(image_path, output_base_dir, conf_threshold=0.25, logger=None):
    """Detect and crop humans with logging"""
    start_time = time.time()

    try:
        model = YOLO('yolov8n.pt')
        results = model(image_path)[0]

        image = cv2.imread(image_path)
        if image is None:
            raise Exception(f"Could not read image: {image_path}")

        output_base_dir = Path(output_base_dir)
        output_base_dir.mkdir(parents=True, exist_ok=True)
        base_filename = Path(image_path).stem

        crops = []
        for idx, detection in enumerate(results.boxes.data):
            x1, y1, x2, y2, conf, cls = detection

            if int(cls) == 0 and conf > conf_threshold:
                x1, y1, x2, y2 = map(int, [x1, y1, x2, y2])
                x1, y1 = max(0, x1), max(0, y1)
                x2, y2 = min(image.shape[1], x2), min(image.shape[0], y2)

                crop = image[y1:y2, x1:x2]
                person_dir = output_base_dir / f"person_{idx}"
                person_dir.mkdir(parents=True, exist_ok=True)

                output_filename = f"{base_filename}_person_{idx}_{conf:.2f}.jpg"
                output_path = person_dir / output_filename

                cv2.imwrite(str(output_path), crop)

                crop_info = {
                    'path': str(output_path),
                    'confidence': float(conf),
                    'bbox': (x1, y1, x2, y2)
                }
                crops.append(crop_info)

        end_time = time.time()

        if logger:
            logger.log_operation("human_detection", {
                "image_path": str(image_path),
                "output_dir": str(output_base_dir),
                "status": "success",
                "duration": end_time - start_time,
                "num_humans": len(crops),
                "crops": crops
            })

        return crops
    except Exception as e:
        if logger:
            logger.log_operation("human_detection", {
                "image_path": str(image_path),
                "output_dir": str(output_base_dir),
                "status": "error",
                "error": str(e)
            })
        raise

if __name__ == "__main__":
    # Initialize logger
    logger = DetectionLogger("detection_logs.json")

    try:
        # Configuration
        image_path = '/content/input/263155-1600x1030-professional-business-attire-women_jpg.rf.0215deaeebf1db466a9245fdd6c65065.jpg'
        output_base_dir = "/content/output/humans"
        model_path = "/content/runs/detect/train3/weights/best.pt"

        # Log configuration
        logger.log_operation("configuration", {
            "image_path": image_path,
            "output_base_dir": output_base_dir,
            "model_path": model_path
        })

        # Detect and crop humans
        print("Detecting and cropping humans...")
        human_crops = detect_and_crop_humans(image_path, output_base_dir, logger=logger)
        print(f"Found {len(human_crops)} humans")

        # Load custom model
        custom_model = load_model(model_path, logger)

        # Process each human crop
        results = []
        for idx, crop_info in enumerate(human_crops):
            print(f"\nProcessing human {idx+1}...")
            human_output_dir = Path(output_base_dir) / f"person_{idx}_detections"

            # Process the cropped image
            try:
                detection_results = process_image(crop_info['path'], custom_model, logger=logger)
                crops = crop_and_save_detections(
                    crop_info['path'],
                    detection_results,
                    human_output_dir,
                    logger=logger
                )

                result = {
                    'image_path': crop_info['path'],
                    'crops': crops
                }
                results.append(result)

            except Exception as e:
                logger.log_operation("crop_processing_error", {
                    "crop_path": crop_info['path'],
                    "error": str(e)
                })
                print(f"Error processing crop {crop_info['path']}: {str(e)}")
                continue

        # Log final summary
        logger.current_session["summary"] = {
            "total_humans_detected": len(human_crops),
            "total_crops_processed": len(results),
            "total_objects_detected": sum(len(r['crops']) for r in results)
        }

        # Save all logs
        logger.save_logs()

        # Print summary
        print(f"\nProcessed {len(results)} human crops")
        for result in results:
            print(f"\nImage: {result['image_path']}")
            print(f"Found {len(result['crops'])} objects")
            for crop in result['crops']:
                print(f"- {crop['class']} (conf: {crop['confidence']:.2f})")

    except Exception as e:
        logger.log_operation("fatal_error", {
            "error": str(e)
        })
        logger.save_logs()
        raise
```

```python
import torch
import torch.nn as nn
import numpy as np
import requests
from PIL import Image
from io import BytesIO
import timm
import torchvision.transforms as transforms

class EfficientNetV2B0Classifier(nn.Module):
    def __init__(self, num_classes):
        super(EfficientNetV2B0Classifier, self).__init__()
        self.model = timm.create_model('tf_efficientnetv2_b2.in1k', pretrained=True)
        num_ftrs = self.model.get_classifier().in_features
        self.model.classifier = nn.Sequential(
            nn.Linear(num_ftrs, num_classes),
            nn.Dropout(0.5)
        )

    def forward(self, x):
        return self.model(x)

def load_trained_model(model_path, num_classes, device='cpu'):
    """
    Load the trained model.
    """
    model = EfficientNetV2B0Classifier(num_classes)
    model.load_state_dict(torch.load(model_path, map_location=device))
    model = model.to(device)
    model.eval()
    return model

def preprocess_image(image, img_size=(224, 224)):
    """
    Preprocess the image for model input.
    """
    transform = transforms.Compose([
        transforms.Resize(img_size),
        transforms.CenterCrop(img_size),
        transforms.ToTensor(),
    ])
    return transform(image).unsqueeze(0)

def predict_image_class(model, image_input, class_names, device='cpu'):
    """
    Predict the class of the image.

    Parameters:
    - model : PyTorch model
    - image_input : Image path, URL, PIL Image, or numpy array
    - class_names : List of class names
    - device : 'cpu' or 'cuda'

    Returns:
    - dict : Predicted class, confidence, and probabilities
    """
    try:
        if isinstance(image_input, str):
            if image_input.startswith(('http://', 'https://')):
                response = requests.get(image_input)
                image = Image.open(BytesIO(response.content)).convert('RGB')
            else:
                image = Image.open(image_input).convert('RGB')
        elif isinstance(image_input, np.ndarray):
            image = Image.fromarray(np.uint8(image_input)).convert('RGB')
        elif isinstance(image_input, Image.Image):
            image = image_input.convert('RGB')
        else:
            raise ValueError("Unsupported input type. Provide an image path, URL, PIL Image, or numpy array.")

        img_tensor = preprocess_image(image).to(device)

        with torch.no_grad():
            outputs = model(img_tensor)
            probabilities = torch.nn.functional.softmax(outputs, dim=1)
            _, predicted = torch.max(outputs.data, 1)
            probs = probabilities.cpu().numpy()[0]

            return {
                'predicted_class': class_names[predicted.item()],
                'confidence': float(probs[predicted.item()]),
                'probabilities': {class_names[i]: float(prob) for i, prob in enumerate(probs)}
            }

    except Exception as e:
        return {'error': str(e)}

def get_top_n_predictions(model, image_input, class_names, n=3, device='cpu'):
    """
    Get top N predictions for an image.

    Parameters:
    - model : PyTorch model
    - image_input : Image path or URL
    - class_names : List of class names
    - n : Number of top predictions to return

    Returns:
    - dict : Top N predictions
    """
    results = predict_image_class(model, image_input, class_names, device)
    if 'error' not in results:
        sorted_probs = sorted(results['probabilities'].items(), key=lambda x: x[1], reverse=True)[:n]
        return {'top_predictions': [{'class': class_name, 'probability': prob} for class_name, prob in sorted_probs]}
    return results

# Example of usage
if __name__ == "__main__":
    # Setup
      import os
import torch
from pathlib import Path
import glob

def load_trained_model(model_path, num_classes, device):
    """Load the trained model"""
    # Your existing model loading code here
    model = None  # Replace with actual model loading
    model = model.to(device)
    model.eval()
    return model

def process_directory(directory_path, model, class_names, device):
    """
    Process all images in the specified directory and make predictions

    Args:
        directory_path (str): Path to directory containing images
        model: Trained PyTorch model
        class_names (list): List of class names
        device (str): Device to use for computation

    Returns:
        dict: Dictionary containing predictions for each image
    """
    # Get all image files in the directory
    image_pattern = os.path.join(directory_path, "person_*_detections", "*.jpg")
    image_files = glob.glob(image_pattern)

    results = {}

    for image_path in image_files:
        try:
            # Get predictions for single image
            single_prediction = predict_image_class(model, image_path, class_names, device)

            # Get top 3 predictions
            top_3 = get_top_n_predictions(model, image_path, class_names, n=3, device=device)

            # Store results
            results[image_path] = {
                'top_prediction': single_prediction,
                'top_3_predictions': top_3
            }

            # Print results for this image
            print(f"\nResults for {os.path.basename(image_path)}:")
            print(f"Top prediction: {single_prediction['predicted_class']} "
                  f"({single_prediction['confidence']:.2%})")
            print("Top 3 predictions:")
            for pred in top_3['top_predictions']:
                print(f"  {pred['class']}: {pred['probability']:.2%}")

        except Exception as e:
            print(f"Error processing {image_path}: {str(e)}")
            results[image_path] = {'error': str(e)}

    return results

if __name__ == "__main__":
    # Setup
    model_path = '/content/drive/MyDrive/Clothing_Data_HRP/Model/final_model_b2_v1.pth'
    num_classes = 8
    class_names = ['female_bodycon', 'female_pant_suit_top', 'female_pants_bottom',
                   'informal cloths men', 'informal cloths women', 'male_pants',
                   'shirt men', 'suits men']

    # Set device
    device = 'cuda' if torch.cuda.is_available() else 'cpu'
    print(f"Using device: {device}")

    # Load model
    try:
        model = load_trained_model(model_path, num_classes, device)
        print("Model loaded successfully")
    except Exception as e:
        print(f"Error loading model: {str(e)}")
        exit(1)

    # Directory containing the images
    base_dir = 'output/humans'

    # Process all images
    print(f"\nProcessing images in {base_dir}...")
    results = process_directory(base_dir, model, class_names, device)

    # Print summary
    print("\nProcessing complete!")
    print(f"Processed {len(results)} images")

    # Count successful vs failed predictions
    successful = sum(1 for r in results.values() if 'error' not in r)
    failed = len(results) - successful
    print(f"Successful predictions: {successful}")
    print(f"Failed predictions: {failed}")

    # Save results to file (optional)
    import json
    output_file = "prediction_results.json"
    with open(output_file, 'w') as f:
        json.dump(results, f, indent=2)
    print(f"\nResults saved to {output_file}")
```

```python
import json
from collections import defaultdict
import re

def combine_person_predictions(input_json):
    """
    Combine predictions based on person detection sections.

    Args:
        input_json (dict): Input JSON with predictions

    Returns:
        dict: Combined predictions grouped by human detection number
    """
    # Dictionary to store combined results
    combined_results = defaultdict(list)

    # Process each image path and its predictions
    for image_path, predictions in input_json.items():
        # Extract person number from the path
        match = re.search(r'person_(\d+)_detections', image_path)
        if not match:
            continue

        person_num = int(match.group(1))

        # Get the predicted class
        predicted_class = predictions['top_prediction']['predicted_class']

        # Add to the combined results
        combined_results[f"human_{person_num}_detection"].append(predicted_class)

        # Keep only top 2 predictions per person
        combined_results[f"human_{person_num}_detection"] = \
            combined_results[f"human_{person_num}_detection"][:2]

    return dict(combined_results)

def process_and_save_predictions(input_path, output_path):
    """
    Process predictions from input JSON and save combined results.

    Args:
        input_path (str): Path to input JSON file
        output_path (str): Path to save combined results
    """
    # Load input JSON
    with open(input_path, 'r') as f:
        input_data = json.load(f)

    # Combine predictions
    combined_results = combine_person_predictions(input_data)

    # Save combined results
    with open(output_path, 'w') as f:
        json.dump(combined_results, f, indent=2)

    return combined_results

if __name__ == "__main__":
    # Example usage
    input_file = "prediction_results.json"
    output_file = "combined_predictions.json"

    results = process_and_save_predictions(input_file, output_file)

    # Print results
    print("\nCombined Predictions:")
    for human_detection, predictions in results.items():
        print(f"\n{human_detection}:")
        print(f"Predicted classes: {predictions}")
```
